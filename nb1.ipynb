```python
from transfer_matrix_method.refractive_index import MaterialRefractiveIndex
from transfer_matrix_method.reflectance import Layer, reflectance
import matplotlib.pyplot as plt
import numpy as np

titanium = MaterialRefractiveIndex("Ti")
cobalt = MaterialRefractiveIndex("Co")
polyethylene = MaterialRefractiveIndex("Polyethylene")



q = MaterialRefractiveIndex("SiO2")
glass = MaterialRefractiveIndex("BK7")
ito =MaterialRefractiveIndex("ITO")
co = MaterialRefractiveIndex("Co")
```

```python
x = np.arange(400e-9,750e-9,1e-9)
layers = [
    Layer(25e-9, ito)
]

y_0 = [reflectance(layers, x_i)["reflectance"] for x_i in x]
y_1 = [reflectance(layers, x_i)["transmittance"] for x_i in x]
# y = [c.refractive_index[x_i]["n"] for x_i in x]

# y = [y_i[0] + abs(y_i[1]) for y_i in y]

fig, ax = plt.subplots()

ax2 = ax.twinx()

ax.plot(x,y_0, label="Reflectance")
ax2.plot(x,y_1, c="red", label="Transmittance")

ax.set_ylabel("Reflectance %")
ax.set_xlabel("Wavelength ($m$)")
ax2.set_ylabel("Transmittance %")

ax.set_title("Transmittance and reflectance across Visible Spectrum")
fig.legend()
fig.tight_layout()
```
