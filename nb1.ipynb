```python
from transfer_matrix_method.refractive_index import CSVRefractiveIndex
from transfer_matrix_method.reflectance import Layer, reflectance
import matplotlib.pyplot as plt
import numpy as np

with open("transfer_matrix_method/data/Co.txt", encoding="utf-8-sig") as f:
    cobalt = CSVRefractiveIndex(f)

with open("transfer_matrix_method/data/Polyethylene.txt", encoding="utf-8-sig") as f:
    polyethelene = CSVRefractiveIndex(f)


p1 = Layer(0.1, polyethelene)
p2 = Layer(0.2, polyethelene)
c = Layer(0.1, cobalt)
```

```python
x = np.arange(380e-9,900e-9,1e-9)
layers = [p1,c,p2]

y = [reflectance(layers, x_i) for x_i in x]
# y = [c.refractive_index[x_i]["n"] for x_i in x]

plt.plot(x,y)
```
